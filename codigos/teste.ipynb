```python
# Carregar o arquivo CSV
jobs_file = 'D:/Mestrado/mineracao-texto/grupo/MinText_Curriculos/teste/vagas.csv'
jobs_df = pd.read_csv(jobs_file)

# Definir uma lista de pontuações e palavras vazias
punctuations = set(string.punctuation)
stop_words = set(stopwords.words('portuguese')) # Dependendo do idioma

# # Função para extrair palavras-chave de uma descrição
# def extract_keywords(description):
#     tokens = word_tokenize(description.lower())
#     clean_tokens = [token for token in tokens if token not in punctuations and token.lower() not in stop_words]
#     return set(clean_tokens)

# # Aplicar a função de extração de palavras-chave à coluna 'description'
# jobs_df['keywords'] = jobs_df['description'].apply(extract_keywords)

# Função para calcular o score de qualidade do texto
# Função para calcular o score de adequação do currículo para uma vaga
# def calculate_resume_score(resume_text, job_description):
#     # Tokenizar e remover palavras e números não relevantes do currículo
#     resume_tokens = [word.upper() for word in word_tokenize(resume_text) if word.isalpha()]
    
#     # Tokenizar e remover palavras e números não relevantes da descrição da vaga
#     job_tokens = [word.upper() for word in word_tokenize(job_description) if word.isalpha()]
    
#     # Calcular a similaridade entre os tokens do currículo e da descrição da vaga
#     similarity = fuzz.token_set_ratio(resume_tokens, job_tokens)
    
#     # Normalizar o valor para o intervalo [0, 1]
#     normalized_score = similarity / 100  
    
#     # Ajustar o score para o intervalo de 0 a 10
#     score = normalized_score * 10
    
#     return score
def calculate_text_quality(generated_text, job_interview):
    tokens = word_tokenize(generated_text.lower())
    clean_tokens = [token for token in tokens if token not in punctuations and token.lower() not in stop_words]
    keyword_count = sum(1 for token in clean_tokens if token in job_interview)
    total_tokens = len(clean_tokens)
    # Verificar se total_tokens é diferente de zero antes de calcular o score
    if total_tokens != 0:
        # Calcular o score no intervalo de 1 a 10 com duas casas decimais
        score = round((keyword_count / total_tokens) * 10, 2)
    else:
        score = 0  # Definir score como zero se total_tokens for zero
    return score
```

```python
# Passo 1: Leitura e Tokenização de Currículos em PDF
def read_pdf(file_path):
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        text = ''
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text

# Passo 2: Carregar Lista de Vagas
def load_jobs(file_path):
    jobs_df = pd.read_csv(file_path)
    return jobs_df

# Passo 3: Associação de Currículos às Vagas
def associate_resumes_to_jobs_fuzzy(resume_folder, jobs_df):
    resume_job_mapping = {}
    for filename in os.listdir(resume_folder):
        if filename.startswith('curriculo') and filename.endswith('.pdf'):
            resume_id = 1
            resume_text = read_pdf(os.path.join(resume_folder, filename))
            best_job_id = 1478
            best_similarity = -1
            for index, job_row in jobs_df.iterrows():
                job_description = job_row['description']
                similarity = fuzz.token_set_ratio(resume_text, job_description)
                if similarity > best_similarity:
                    best_similarity = similarity
                    best_job_id = job_row['id']
            resume_job_mapping[resume_id] = best_job_id
    return resume_job_mapping

# Passo 4: Treinamento do Modelo SVM
def train_svm(resumes, job_labels):
    X_train, X_test, y_train, y_test = train_test_split(resumes, job_labels, test_size=0.2, random_state=42)
    svm_model = SVC()
    svm_model.fit(X_train, y_train)
    return svm_model

# Passo 5: Avaliação do Modelo SVM
def evaluate_svm(model, X_test, y_test):
    predictions = model.predict(X_test)
    report = classification_report(y_test, predictions)
    print(report)

# Passo 7: Avaliação dos Currículos pela LLM
# def evaluate_with_gpt3(resumes, jobs):
#     evaluation_results = []
#     for resume_id, job_id in resumes.items():
#         resume_text = resumes[resume_id]
#         job_row = jobs[jobs['id'] == 1478]
#         if not job_row.empty:
#             job_description = job_row['description'].iloc[0]
#             job_id = job_row['id'].iloc[0]
#             question = f"Qual é a sua experiência relevante para a posição de {job_description}?"
#             suggested_response = f"Candidato: {resume_text} Vaga: {job_description}"
#             # Chamar a API do GPT-3 para gerar a resposta
#             response = openai.Completion.create(
#                 engine="text-davinci-003",  # Especificar o modelo do GPT-3
#                 prompt=f"Question: {question}\nAnswer:",
#                 max_tokens=150,  # Limite de tokens para a resposta
#                 temperature=0.7,  # Controla a aleatoriedade da amostragem
#                 stop="\n"  # Parar quando encontrar uma nova linha
#             )
#             generated_text = response.choices[0].text.strip()
#             print(generated_text)
#             # Avaliar a qualidade do texto gerado
#             score = calculate_text_quality(generated_text, job_description)
#             evaluation_results.append({'resume_id': resume_id, 'job_id': job_id, 'score': score})
#         else:
#             print(f"Vaga com ID {job_id} não encontrada.")
#     return evaluation_results
# MAX_LENGTH = 512  # Tamanho máximo para o texto de entrada, de acordo com a capacidade do modelo BERT

# def evaluate_with_bert(resumes, jobs):
#     evaluation_results = []
#     for resume_id, job_id in resumes.items():
#         resume_text = resumes[resume_id]
#         job_row = jobs[jobs['id'] == job_id]
#         if not job_row.empty:
#             job_description = job_row['description'].iloc[0]
#             # Concatenar a pergunta com o currículo
#             question = f"Qual é a sua experiência relevante para a posição de {job_description}?"
#             context = f"Candidato: {resume_text} Vaga: {job_description}"
#             # Limitar o comprimento do texto de entrada
#             inputs_text = question + " " + context
#             inputs_text = inputs_text[:MAX_LENGTH]
#             # Tokenizar e codificar o texto de entrada
#             inputs = tokenizer(inputs_text, return_tensors="pt", max_length=MAX_LENGTH, truncation=True)
#             try:
#                 # Executar o modelo BERT
#                 outputs = model(**inputs)
#                 # Encontrar a melhor resposta
#                 answer_start_scores = outputs.start_logits
#                 answer_end_scores = outputs.end_logits
#                 # Selecionar os índices da resposta com maior probabilidade
#                 answer_start = torch.argmax(answer_start_scores)
#                 answer_end = torch.argmax(answer_end_scores) + 1
#                 # Verificar se a resposta é válida
#                 input_ids = inputs["input_ids"][0]
#                 if answer_start < answer_end and answer_end <= len(input_ids):
#                     # Converter os tokens da resposta em texto
#                     answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end]))
#                     # Avaliar a qualidade da resposta
#                     score = calculate_text_quality(answer, job_description)
#                     evaluation_results.append({'resume_id': resume_id, 'job_id': job_id, 'score': score})
#                 else:
#                     print("Erro: Resposta inválida")
#             except Exception as e:
#                 print(f"Erro ao processar currículo {resume_id}: {str(e)}")
#         else:
#             print(f"Vaga com ID {job_id} não encontrada.")
#     return evaluation_results

def evaluate_with_llm_fuzzy(resumes, jobs):
    evaluation_results = []
    for resume_id, job_id in resumes.items():
        resume_text = resumes[resume_id]
        job_row = jobs[jobs['id'] == 1478]
        if not job_row.empty:
            job_description = job_row['description'].iloc[0]
            job_id = job_row['id'].iloc[0]
            question = f"Qual é a sua experiência relevante para a posição de {job_description}?"
            desired_response = f"Para a posição de {job_description}, minha experiência como {resume_text} me qualifica porque..."
            # Combinação da pergunta e resposta desejada
            interview_text = f"Pergunta: {question}\nResposta: {desired_response}"
            # Executar a LLM
            generated_text = llm_interviewer(interview_text, max_length=200, num_return_sequences=1)[0]['generated_text']
            print(generated_text)
            # Avaliar a qualidade do texto gerado
            score = calculate_text_quality(generated_text, job_description)
            evaluation_results.append({'resume_id': resume_id, 'job_id': job_id, 'score': score})
        else:
            print(f"Vaga com ID {job_id} não encontrada.")
    return evaluation_results

# Passo 8: Geração de Novo Arquivo CSV
def generate_csv(results, output_file):
    with open(output_file, 'w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=['resume_id', 'job_id', 'score'])
        writer.writeheader()
        writer.writerows(results)
```

```python
# Função para calcular o score de adequação do currículo para uma vaga
def calculate_resume_score(resume_text, job_description):
    # Tokenizar e remover palavras e números não relevantes do currículo
    resume_tokens = [word.upper() for word in word_tokenize(resume_text) if word.isalpha()]
    
    # Tokenizar e remover palavras e números não relevantes da descrição da vaga
    job_tokens = [word.upper() for word in word_tokenize(job_description) if word.isalpha()]
    
    # Calcular a similaridade entre os tokens do currículo e da descrição da vaga
    similarity = fuzz.token_set_ratio(resume_tokens, job_tokens)
    
    # Normalizar o valor para o intervalo [0, 1]
    normalized_score = similarity / 100  
    
    # Ajustar o score para o intervalo de 0 a 10
    score = normalized_score * 10
    
    return score

# Passo 7: Avaliação dos Currículos
def evaluate_resumes(resumes_folder, jobs_df):
    evaluation_results = []
    for filename in os.listdir(resumes_folder):
        if filename.startswith('1002543') and filename.endswith('.pdf'):
            resume_id = os.path.splitext(filename)[0]
            resume_text = read_pdf(os.path.join(resumes_folder, filename))
            for index, job_row in jobs_df.iterrows():
                job_id = job_row['id']
                job_description = job_row['description']
                score = calculate_resume_score(resume_text, job_description)
                evaluation_results.append({'resume_id': resume_id, 'job_id': job_id, 'score': score})
    return evaluation_results

# Avaliação dos currículos
evaluation_results = evaluate_resumes(resume_folder, jobs_df)

# Geração de novo arquivo CSV
output_csv = 'resultado.csv'
generate_csv(evaluation_results, output_csv)

# Visualizar o resultado
print(pd.read_csv('resultado.csv'))
```
